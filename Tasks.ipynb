This first task is to create a dictionary from a given list, and to generate an accompanying value for each unique element that states the number of times that this element appears in the given list.

```python
# First create a list of values:
A = [10, 20, 10, 30, 20, 10, 50, 50, 40, 10, 40]
```

[1]:ThisPointer.com: How to convert a list to dictionary: https://thispointer.com/python-how-to-convert-a-list-to-dictionary/

```python
# Next step is to find the unique values within the list:
unique_values_A=[]
for i in A:
    if i not in unique_values_A:
        unique_values_A.append(i)
print(unique_values_A)
```

The next step would seem to be to figure out how to count the number of instances of a given element in the list:

```python
# Next we find a way of counting an element in the list:
count10 = 0

for i in A:
    if(i == 10):
     count10 = count10 + 1
print ("{}:{}".format(10, count10))
```

So from here we can see that there is an obvious initial solution to the problem, which is to repeat the above step for each unique element. 

```python
# Repeaet last step for each element

count10 = 0
count20 = 0
count30 = 0
count40 = 0
count50 = 0

for i in A:
    if(i == 10):
     count10 = count10 + 1
print ("{}:{}".format(10, count10))

for i in A:
    if(i == 20):
     count20 = count20 + 1
print ("{}:{}".format(20, count20))

for i in A:
    if(i == 30):
     count30 = count30 + 1
print ("{}:{}".format(30, count30))

for i in A:
    if(i == 40):
     count40 = count40 + 1
print ("{}:{}".format(40, count40))

for i in A:
    if(i == 50):
     count50 = count50 + 1
print ("{}:{}".format(50, count50))
```

I would imagine, however, that this is the most tedious possible approach, and in the case of a larger list with more elements this might begin to become near impossible, and would also result in an extremely long piece of code.

From a bit more trawling of a variety of forums and tutorials I have found a way of creating a dictionary that gets us up to the next point, in which from just a few lines of code we have a dictionary of key elements, accompanied by an empty string. The next step is finding out how to pair the counted values from the above function to the key elements.

```python
# Create empty dictionary
# populate dictionary with key elements and accompanying empty strings using for loop & '.update'

converted_dict = {}
for i in A:
    converted_dict.update({i:""})
print(converted_dict)
```

### Solution:

```python
# Add A.count(i) to above function:
converted_dict = {}
for i in A:
    converted_dict.update({i:A.count(i)})
print(converted_dict)
```

'Simpson' Paradox' is a term used in data analytics for a phenomenon that may occur when two or more similar datasets are combined, in which the original trends applied to both cause the new combined trend to become severely altered, or often fully inverted. Here we will explore why this occurs, what it means, and what it's implications may be in real life.

[14] Wikepedia: Simpson's paradox: https://en.wikipedia.org/wiki/Simpson%27s_paradox

[15] Britanica.org: Simpson's Paradox Explained: https://www.britannica.com/topic/Simpsons-paradox

[16] BMC Medical Research Methodology: https://bmcmedresmethodol.biomedcentral.com/articles/10.1186/1471-2288-8-34

The best way to illustrate the phenomenon is to create four sepperate scatter plots representing similar but different datasets. We can use the function of a line with added noise to reverse the concept of linear regression and simulate similar data sets of given trends. We will then combine the data sets and perform linear regression on the combined scatter plot in order to visualise the new trend created.

```python
# create some x values:
x1 = np.linspace(0.0, 10.0, 100)
x2 = np.linspace(2.0, 12.0, 100)
x3 = np.linspace(4.0, 14.0, 100)
x4 = np.linspace(6.0, 16.0, 100)
```

```python
# create some y values:
y1 = 2.0 * x1 + 1
y2 = 4.0 * x2 + 2
y3 = 6.0 * x3 + 3
y4 = 8.0 * x4 + 4
```

```python
# plot first dataset
plt.plot(x1, y1)
```

Here we have plotted 100 points in a data set, all of which form a straight line. Imagine that this was the line given by fitting the best possible line to a given data set. This would then be the 'trend' of the dataset. How then, can we simulate what that original dataset might look like? We can use noise to randomize the placement of the dots to a given degree so that the data looks like it may have been the origin of this trend.
