## Task 1: Counts.

This first task is to create a dictionary from a given list, and to generate a value for each unique element that states the number of times that this element appears in the given list. 

```python
# First create a list of values:
A = [10, 20, 10, 30, 20, 10, 50, 50, 40, 10, 40]
```

```python
# Next step is to find the unique values withing the list:
unique_values_A=[]
for i in A:
    if i not in unique_values_A:
        unique_values_A.append(i)
print(unique_values_A)
```
