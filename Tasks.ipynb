This first task is to create a dictionary from a given list, and to generate an accompanying value for each unique element that states the number of times that this element appears in the given list.

```python
# First create a list of values:
A = [10, 20, 10, 30, 20, 10, 50, 50, 40, 10, 40]
```

[1]:ThisPointer.com: How to convert a list to dictionary: https://thispointer.com/python-how-to-convert-a-list-to-dictionary/

```python
# Next step is to find the unique values within the list:
unique_values_A=[]
for i in A:
    if i not in unique_values_A:
        unique_values_A.append(i)
print(unique_values_A)
```

[2]: Infopedic Techie: Find the unique values in a list: https: //www.youtube.com/watch?v=7f2UJgig2yI

<br>

The next step would seem to be to figure out how to count the number of instances of a given element in the list:

```python
# Next we find a way of counting an element in the list:
count10 = 0

for i in A:
    if(i == 10):
     count10 = count10 + 1
print ("{}:{}".format(10, count10))
```

[3]: SDET - QA Automation Techie: Frequently Asked Python Program 15: Count Occurrences of an element in a list: https://www.youtube.com/watch?v=Fc6OOmf3Hcs

<br>

So from here we can see that there is an obvious initial solution to the problem, which is to repeat the above step for each unique element. 

```python
# Repeaet last step for each element

count10 = 0
count20 = 0
count30 = 0
count40 = 0
count50 = 0

for i in A:
    if(i == 10):
     count10 = count10 + 1
print ("{}:{}".format(10, count10))

for i in A:
    if(i == 20):
     count20 = count20 + 1
print ("{}:{}".format(20, count20))

for i in A:
    if(i == 30):
     count30 = count30 + 1
print ("{}:{}".format(30, count30))

for i in A:
    if(i == 40):
     count40 = count40 + 1
print ("{}:{}".format(40, count40))

for i in A:
    if(i == 50):
     count50 = count50 + 1
print ("{}:{}".format(50, count50))
```

I would imagine, however, that this is the most tedious possible approach, and in the case of a larger list with more elements this might begin to become near impossible, and would also result in an extremely long piece of code.

From a bit more trawling of a variety of forums and tutorials I have found a way of creating a dictionary that gets us up to the next point, in which from just a few lines of code we have a dictionary of key elements, accompanied by an empty string. The next step is finding out how to pair the counted values from the above function to the key elements.

```python
# Create empty dictionary
# populate dictionary with key elements and accompanying empty strings using for loop & '.update'

converted_dict = {}
for i in A:
    converted_dict.update({i:""})
print(converted_dict)
```

[4]:medium.com: Create a dictionary from a list: https://medium.com/swlh/create-a-dictionary-from-a-list-65742246ab4b

<br>

(I will just make a brief statement here to clarify that up until this point I had been labouring under the illusion that certain basic operators, (namely 'count' & 'set') were out of bounds for this module, and as a result had been trying to re-invent the wheel of sorts. I had been struggling to fit these alternative functions to fit the solution in an inumerate variety of failed permutations, but now that I have clarified that these operators do not fall under the criteria of "any module from the standard library1 or otherwise", I think the solution should be a bit more obvious. I will, however, leave the initial functions at least, in case there is any debate as to whether or not I suffered enough coming to a solution.)
***

### Solution:

```python
# Add A.count(i) to above function:
converted_dict = {}
for i in A:
    converted_dict.update({i:A.count(i)})
print(converted_dict)
```

First create a random number generateor that can simulate a single 6 sided die:

As specified in the description of the task, create two variables, 'k' (Number of Dice) and 'n'(Number of rolls), and use these variables to generate an array of outcomes for each roll:

As the aim is to focus on the sum of the outcomes, we have used the 'sum' opperator to add the outcome of any one set of dice being rolled together.

```python
# The values 'k' & 'n' below can be changed to gain different results. 

# Create a varable 'k' to signify number of dice:
k = 7

# and a variable 'n' for number of times these dice are rolled
n = 500

# Input variables into die rng and add total of each roll together:
# This will generate an array of total values of all the dice rolled at any one time.

dicerolls = sum(rng.integers(1, 6, (k, n)))
dicerolls
```

```python
# Convert resultant array to a list so that it can be converted into a dictionary.

arr = dicerolls
print(f'NumPy Array:\n{arr}')

list1 = arr.tolist()
print(f'List: {list1}')
```
