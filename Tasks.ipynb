This first task is to create a dictionary from a given list, and to generate an accompanying value for each unique element that states the number of times that this element appears in the given list.

```python
# First create a list of values:
A = [10, 20, 10, 30, 20, 10, 50, 50, 40, 10, 40]
```

```python
# Next step is to find the unique values within the list:
unique_values_A=[]
for i in A:
    if i not in unique_values_A:
        unique_values_A.append(i)
print(unique_values_A)
```

The next step would seem to be to figure out how to count the number of instances of a given element in the list:

```python
# Next we find a way of counting an element in the list:
count10 = 0

for i in A:
    if(i == 10):
     count10 = count10 + 1
print ("{}:{}".format(10, count10))
```

So from here we can see that there is an obvious initial solution to the problem, which is to repeat the above step for each unique element. 

```python
# Repeaet last step for each element

count10 = 0
count20 = 0
count30 = 0
count40 = 0
count50 = 0

for i in A:
    if(i == 10):
     count10 = count10 + 1
print ("{}:{}".format(10, count10))

for i in A:
    if(i == 20):
     count20 = count20 + 1
print ("{}:{}".format(20, count20))

for i in A:
    if(i == 30):
     count30 = count30 + 1
print ("{}:{}".format(30, count30))

for i in A:
    if(i == 40):
     count40 = count40 + 1
print ("{}:{}".format(40, count40))

for i in A:
    if(i == 50):
     count50 = count50 + 1
print ("{}:{}".format(50, count50))
```

I would imagine, however, that this is the most tedious possible approach, and in the case of a larger list with more elements this might begin to become near impossible, and would also result in an extremely long piece of code.

From a bit more trawling of a variety of forums and tutorials I have found a way of creating a dictionary that gets us up to the next point, in which from just a few lines of code we have a dictionary of key elements, accompanied by an empty string. The next step is finding out how to pair the counted values from the above function to the key elements.

```python
# Create empty dictionary
# populate dictionary with key elements and accompanying empty strings using for loop & '.update'

converted_dict = {}
for i in A:
    converted_dict.update({i:""})
print(converted_dict)
```

### Solution:

```python
# Add A.count(i) to above function:
converted_dict = {}
for i in A:
    converted_dict.update({i:A.count(i)})
print(converted_dict)
```

'Simpson' Paradox' is a term used in data analytics for a phenomenon that may occur when two or more similar datasets are combined, in which the original trends applied to both cause the new combined trend to become severely altered, or often fully inverted. Here we will explore why this occurs, what it means, and what it's implications may be in real life.

[14] Wikepedia: Simpson's paradox: https://en.wikipedia.org/wiki/Simpson%27s_paradox

[15] Britanica.org: Simpson's Paradox Explained: https://www.britannica.com/topic/Simpsons-paradox

[16] BMC Medical Research Methodology: https://bmcmedresmethodol.biomedcentral.com/articles/10.1186/1471-2288-8-34

The best way to illustrate the phenomenon is to create four sepperate scatter plots representing similar but different datasets. We can use the function of a line with added noise to reverse the concept of linear regression and simulate similar data sets of given trends. We will then combine the data sets and perform linear regression on the combined scatter plot in order to visualise the new trend created.

```python
# create some x values:
x1 = np.linspace(0.0, 10.0, 100)
x2 = np.linspace(2.0, 12.0, 100)
x3 = np.linspace(4.0, 14.0, 100)
x4 = np.linspace(6.0, 16.0, 100)
```

```python
# create some y values:
y1 = 2.0 * x1 + 1
y2 = 4.0 * x2 + 2
y3 = 6.0 * x3 + 3
y4 = 8.0 * x4 + 4
```

```python
# plot first dataset
plt.plot(x1, y1)
```

Here we have plotted 100 points in a data set, all of which form a straight line. Imagine that this was the line given by fitting the best possible line to a given data set. This would then be the 'trend' of the dataset. How then, can we simulate what that original dataset might look like? We can use noise to randomize the placement of the dots to a given degree so that the data looks like it may have been the origin of this trend.

```python
# Add noise to our create some y values:
y1 = 2.0 * x1 + 1 + np.random.normal(0.0, 2, len(x1))
y2 = 4.0 * x2 + 2 + np.random.normal(0.0, 3, len(x2))
y3 = 6.0 * x3 + 3 + np.random.normal(0.0, 1, len(x3))
y4 = 8.0 * x4 + 4 + np.random.normal(0.0, 2, len(x4))
```

```python
# plot first dataset with added noise:
plt.plot(x1, y1)
```

Here we get a line, so let's add dots instead:

```python
# plot first dataset with added noise:
plt.plot(x1, y1, 'o')
```

Now, we know of course that the realtionship between x1 and y1 is characterised by the equation:

- 'y1 = 2.0 * x1 + 1 + np.random.normal(0.0, 2, len(x1))'

or in other words: y = mx + c (plus noise of course), with m & c being the coefficients (or parameters) of the equation, x^1 and x^0.

What if we then asked numpy.polyfit to derive what it percieves to be the best trending line for these dots to be then:

```python
# plot new coefficients based on noise generated dots:
coeffs1 = np.polyfit(x1, y1, 1)
coeffs1
```

```python
# plot the best fit line over the data points:
plt.plot(x1, y1, 'o', label="Data")
plt.plot(x1, coeffs1[0] * x1 + coeffs1[1], '-', label="Best fit")
plt.legend();
```

Now we can plot new coefficients for each of our other datasets:

```python
# plot new coefficients for other y values:
coeffs2 = np.polyfit(x2, y2, 1)
coeffs2
```

```python
# plot new coefficients:
coeffs3 = np.polyfit(x3, y3, 1)
coeffs3
```

```python
# plot new coefficients:
coeffs4 = np.polyfit(x4, y4, 1)
coeffs4
```

And thus we can plot each data set in the same way as our first:

```python
# plot the best fit line over the data points:
plt.plot(x2, y2, 'o', label="Data2")
plt.plot(x2, coeffs2[0] * x2 + coeffs2[1], '-', label="Best fit2")
plt.legend();
```

```python
# plot the best fit line over the data points:
plt.plot(x3, y3, 'o', label="Data3")
plt.plot(x3, coeffs3[0] * x3 + coeffs3[1], '-', label="Best fit3")
plt.legend();
```

```python
# plot the best fit line over the data points:
plt.plot(x4, y4, 'o', label="Data4")
plt.plot(x4, coeffs4[0] * x4 + coeffs4[1], '-', label="Best fit4")
plt.legend();
```

Now that we have all four data sets, we can combine their dat and plot them to the same graph. We can then see what affect this combining of data has on an overall best fir line. It is important to note that although the trends may be similar, they are not as closely related as they may seem in their respective graphs, as each plot has generated its own x and y axis to best represent that particular dataset, and thus the values represented are different respectively.

```python
x = np.concatenate([x1, x2, x3, x4])
```

```python
y = np.concatenate([y1, y2, y3, y4])
```

We can now find the coefficients of the combined datasets:

```python
coeffs = np.polyfit(x, y, 1)
coeffs
```

```python
# plot all 4 data sets, and combined data set best line:
plt.plot(x1, y1, 'o', label="Data")
plt.plot(x1, coeffs1[0] * x1 + coeffs1[1], '-', label="Best fit")
plt.plot(x2, y2, 'o', label="Data2")
plt.plot(x2, coeffs2[0] * x2 + coeffs2[1], '-', label="Best fit2")
plt.plot(x3, y3, 'o', label="Data3")
plt.plot(x3, coeffs3[0] * x3 + coeffs3[1], '-', label="Best fit3")
plt.plot(x4, y4, 'o', label="Data4")
plt.plot(x4, coeffs4[0] * x4 + coeffs4[1], '-', label="Best fit4")
plt.plot(x, y, 'o', label="Data Combined")
plt.plot(x, coeffs[0] * x + coeffs[1], '-', label="Best fit Combined")
plt.legend();
```

As we can see, the best fit line for both doesn't really match any at all. If these lines were all parrallell, the best fit line would be the line perpendicular to all 4, meaniing it would intercept all four, but misrepresent all of them a any point other than te intercept. It is not as extreme here, but if you look at how independant the new line is from any of the dataset, it really does become obvious that it cannot be trusted to properly represent any of them.

So what are the implications of this experiment? Well, it proves that mixing the wrong data, even seemingly similar data, can generate serious misrepresentations within an analysis. 
<br>

Not only can this lead to mistakes, it can also be utilised for more nefarious purposes, in which data can be skewed purposefully in order to achieve a desired result, rather than a true one.
<br>

This Statistical bias is described in the papers we studied in this course. It is explicitly mentioned in the Berkley study of admissions biases. If you consider how the pooling of all departents into one data pool skewed the results of individual departments admissions data, you could argue that that is a perfect example of Simpsons Paradox.
<br>

According to another study by Howard Wainer (listed below) regarding the test scores of different states, the test results are again skewed when pooled together. You could argue that because the average test score of a state is higher than that of another that it has a better education system, however this may not be the case. If factors such as socio-economic and racial factors are taken into account, we can say that a pool of mostly privelaged students would be expected to score higher, and less privelaged groups lower. If a state has more underprivelaged students, but their collecive test scores are higher than their underprivelaged counterparts in other states, the education system there must be doing a better job. Howevever, if said counterparts are mixed with a much larger pool of even mediocre privelaged students then the result may be skewed in favor of a state that is relying more on socio-economic privelages than it is on quality of education to raise its average.
